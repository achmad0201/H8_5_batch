```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv(r"C:\Users\Adi\Documents\codingan\melb_data.csv")
df.head()
```

### Soal no 1

```python
plt.style.use('seaborn')
plt.figure(figsize=(7,5))
plt.hist(df["Price"])
plt.ylabel('Count')
plt.xlabel('Price')
plt.title("Distribusi Harga Rumah", size=15)
plt.show()
```

### Soal no 2

```python
sns.boxplot(x = df["Distance"], y = df['Regionname'])
plt.title("Distribusi Jarak ke CBD berdasarkan Nama Region", size=15)
plt.show()
```

### Soal no 3

```python
sns.barplot(data=df, x = "Type", y = "Price")
plt.title("Perbandingan Nilai Median Harga Rumah", size=15)
plt.ylabel('Price')
plt.xlabel('Type')
plt.show()
```

### Soal no 4

```python
sns.scatterplot(data=df, x = "Price", y = "Distance", hue = 'Type')
plt.title("Korelasi Harga Rumah dan Jarak Rumah ke CBD", size=15)
plt.ylabel('Distance')
plt.xlabel('Price')
plt.show()
```

### Soal no 5

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=False)
```

```python
df['Month'] = df['Date']
```

```python
sns.lineplot(data=df, x = 'Month', y = "Price", hue = 'Type')
plt.title("Tren Harga Rumah per Bulan berdasarkan Tipe Rumah", size=15)
plt.ylabel('Distance')
plt.xlabel('Month')
plt.show()
```
