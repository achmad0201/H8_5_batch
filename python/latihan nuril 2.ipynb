### Soal no 5

```python
df['Status BMI'] = pd.cut(
    df['BMI'], bins=[0, 18, 24.9, 29.9, float('inf')],
    labels=['Underweight', 'Normal', 'Overweight', 'Obesity']
)
df
```

### cara 2

```python
df["Status BMI"] = ''

for i in range(len(df)):
  if df['BMI'][i] >= 18.5 and df['BMI'][i] <= 24.9:
    df["Status BMI"][i] = 'Normal'
  elif df['BMI'][i] >= 25 and df['BMI'][i] <= 29.9:
    df["Status BMI"][i] = 'Overweight'
  elif df['BMI'][i] >= 30 and df['BMI'][i] <= 34.9:
    df["Status BMI"][i] = 'Obesity'
  elif df['BMI'][i] < 18.5:
    df["Status BMI"][i] = 'Underweight'
  else:
    df["Status BMI"][i] = 'Error'
df
```
